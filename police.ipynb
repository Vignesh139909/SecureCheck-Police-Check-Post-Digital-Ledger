```python
import pandas as pd
path=r"C:\Users\Vignesh\OneDrive\Desktop\Datascience\PROJECT1\Police_logs\traffic_stops - traffic_stops_with_vehicle_number.csv"
police_data=pd.read_csv(path)
police_data.head(10)
```

```python
df=police_data.copy()
df.head(10)
```

```python
df.shape
```

```python
df.info()
```

```python
# Data Cleaning and Preprocessing:

df.columns = df.columns.str.strip()

df['stop_date'] = pd.to_datetime(df['stop_date'], format='%Y-%m-%d',errors='coerce')

df['driver_gender'] = df['driver_gender'].replace({ 'M': 'Male', 'F': 'Female' })

df.loc[df['search_conducted'] == False, 'search_type'] = 'No Search'

df.fillna({'search_type': 'Unknown'}, inplace=True)

df.rename(columns={'country_name': 'country'}, inplace=True)

new_order = [
    "country","vehicle_number","stop_date", "stop_time","driver_gender", 
    "driver_age", "driver_race", "violation", "search_conducted", "search_type",
    "stop_outcome", "is_arrested","drugs_related_stop","stop_duration","driver_age_raw","violation_raw"
]

df = df[new_order]


df.head(10)
```

```python
import mysql.connector

try:
    mydb=mysql.connector.connect(
        host="localhost",
        user="root",
        password=""
    )
    mycursor=mydb.cursor()
    print("✅ Sql connection established successfully")
except mysql.connector.Error as err:
    print(f"❌ ERROR IN CONNECTION: {err}")
```

```python
mycursor.execute("DROP DATABASE IF EXISTS Securecheck_ledger")
mycursor.execute("CREATE DATABASE IF NOT EXISTS Securecheck_ledger")
print("✅ Database created successfully")
```

```python
mycursor.execute("USE Securecheck_ledger")
print("✅ Database selected successfully")
```

```python
mycursor.execute("""
    CREATE TABLE IF NOT EXISTS police_stops (
        country VARCHAR(50),
        vehicle_number VARCHAR(20),
        stop_date DATE,
        stop_time TIME,
        driver_gender VARCHAR(10),
        driver_age INT,
        driver_race VARCHAR(10),
        violation VARCHAR(50),
        search_conducted BOOLEAN,
        search_type VARCHAR(50),
        stop_outcome VARCHAR(50),
        is_arrested BOOLEAN,
        drugs_related_stop BOOLEAN,
        stop_duration VARCHAR(20),
        driver_age_raw INT,
        violation_raw VARCHAR(50)
    )
""")

print("✅ Table created successfully")
```

```python
from sqlalchemy import create_engine
engine = create_engine('mysql+mysqlconnector://root:@localhost/Securecheck_ledger')
try:
    df.to_sql('police_stops', con=engine, if_exists='replace', index=False,chunksize=1000)
    print("✅ Data inserted successfully.")
except Exception as e:
    print("❌ Error inserting data:", e)
```
